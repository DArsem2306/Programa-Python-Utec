### **Paso 0:** Cargar librerías y seleccionar el conjunto de datos que se utilizará para este assignment.

```python
#Importar Librerias
import pandas as pd
import seaborn as sns
import random
```

La función `cargarDatos()` escogerá al azar el conjunto de datos (dataset) con el que trabajarás en este assignment. Además la función ya te devuelve el dataset cargado en la variable `df`

```python
def cargarDatos():
  datasetList = ["taxis", "flights", "planets"]
  rand_idx = random.randrange(len(datasetList))
  random_dataset = datasetList[rand_idx]
  print("El dataset que utilizarás es: ", random_dataset)
  df = sns.load_dataset(random_dataset)
  return df
```

```python
#Asigna el dataset al azar y carga el dataset
df= cargarDatos()
df.head()
```

### **Paso 1:** Descubre/Explora el conjunto de datos

```python
#Resumen del dataset
df.info()
```

### **Ejercicio 1:** Método gráfico de un dato categórico ✋

Analiza de manera gráfica un dato categórico de tu dataset.

```python
#Validacion de data por meses
sns.countplot(data=df, x='month', order=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
```

**La interpretación de tu gráfico anterior:**

El gráfico muestra que todos los meses tienen la misma cantidad de registros (12), lo que indica que el dataset tiene datos completos para cada mes del año.

### **Ejercicio 2:** Método gráfico de un dato cuantitativo ✋

Analiza de manera gráfica un dato cuantitativo de tu dataset.

```python
#Historia de Pasajeros por vuelo
sns.histplot(data=df, x='passengers', bins=20, kde=True)
```

**La interpretación de tu gráfico anterior:** ------


La mayoría de los meses tienen entre hasta 250 pasajeros, con una menor cantidad de meses con valores extremos. La distribución tiene una cola hacia la derecha, indicando crecimiento en el tiempo. Esto sugiere que el numero de pasajeros mayores a 300 ocurrieron con menor frecuencia, probablemente en años posteriores.

### **Ejercicio 3**: Método gráfico multivariado para datos cuantitativos ✋


```python
#Heatmap de pasajeros por Año y Mes transcurrido
pivot_df = df.pivot(index="year", columns="month", values="passengers")
sns.heatmap(pivot_df, cmap="YlGnBu", annot=True, fmt="d")
```

**La interpretación de tu gráfico anterior:** ------


A lo largo de los años, los colores del heatmap se intensifican, lo que refleja un crecimiento sostenido en el número de pasajeros desde 1949 hasta 1960. Esto sugiere una expansión de la industria aérea durante la década. Además, el mapa de calor muestra que los meses de julio y agosto, tienen consistentemente un mayor número de pasajeros cada año evidenciando una estacionalidad marcada, posiblemente relacionada a vacaciones.
